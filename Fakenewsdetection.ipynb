```python
import os
import pandas as pd

def load_text_files_from_folder(folder_path, label):
    texts = []
    for file in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file)
        with open(file_path, 'r', encoding='utf-8') as f:
            text = f.read()
            texts.append(text)
    return pd.DataFrame({'text': texts, 'label': label})

fake_folder = '/content/Hindi_fake_news/Hindi_fake_news'
real_folder = '/content/Hindi_real_news/Hindi_real_news'

# Science/health fake news examples
science_fake_samples = [
    "वैज्ञानिकों ने एक ऐसी चमत्कारी दवा खोजने का दावा किया है जिससे सभी बीमारियों का इलाज सिर्फ एक घंटे में संभव है। यह खबर पूरी तरह से झूठी बताई जा रही है।",
    "एक वायरल पोस्ट में कहा गया है कि अब कैंसर का इलाज घर पर बैठकर किया जा सकता है। विशेषज्ञों ने इस खबर को अफवाह करार दिया है।",
    "सोशल मीडिया पर फैली खबर के अनुसार एक नई वैक्सीन से सभी बीमारियां हमेशा के लिए खत्म हो जाएंगी। स्वास्थ्य मंत्रालय ने इस खबर का खंडन किया है।",
    "व्हाट्सएप पर एक मैसेज वायरल हो रहा है जिसमें दावा किया गया है कि सिर्फ पानी पीकर कोरोना का इलाज किया जा सकता है। यह खबर पूरी तरह से झूठी है।",
    "खबर है कि वैज्ञानिकों ने एक ऐसी मशीन बनाई है जो किसी भी वायरस को एक मिनट में खत्म कर सकती है। इस दावे की कोई पुष्टि नहीं हुई है।",
    "एक अफवाह के अनुसार अब से हर नागरिक को मुफ्त दवा किट दी जाएगी जिससे किसी भी बीमारी का इलाज संभव होगा। यह खबर झूठी है।",
    "सोशल मीडिया पर दावा किया गया है कि चंद्रमा पर जीवन की पुष्टि हो गई है और वहां की मिट्टी से कैंसर का इलाज संभव है। यह खबर गलत है।",
    "एक मैसेज में कहा गया है कि मोबाइल फोन की किरणें कोरोना वायरस को मार सकती हैं। वैज्ञानिकों ने इस दावे को गलत बताया है।",
    "वायरल पोस्ट में दावा है कि सूरज की किरणों से किसी भी बीमारी का इलाज हो सकता है। विशेषज्ञों ने इस खबर को अफवाह बताया है।",
    "खबर है कि अब से किसी भी बीमारी के लिए डॉक्टर की जरूरत नहीं पड़ेगी क्योंकि एक नई चाय पत्ती से इलाज संभव है। यह पूरी तरह से झूठा दावा है।"
]

# Official-style fake news examples
official_fake_samples = [
    "सरकार अगले महीने से हर नागरिक को मुफ्त घर देने जा रही है। इस खबर की कोई आधिकारिक पुष्टि नहीं हुई है।",
    "व्हाट्सएप पर फैली एक खबर के अनुसार अब से पेट्रोल मुफ्त मिलेगा। सरकार ने इस पर कोई टिप्पणी नहीं की है।",
    "सोशल मीडिया पर वायरल खबर में दावा किया गया है कि सरकार हर छात्र को मुफ्त लैपटॉप देने जा रही है। यह खबर गलत है।",
    "एक वायरल पोस्ट में कहा जा रहा है कि अब से सभी बीमारियों का इलाज सिर्फ एक दवा से संभव है। विशेषज्ञों ने इसे अफवाह बताया है।",
    "खबर है कि अगले साल से सभी छात्रों को बिना परीक्षा के पास कर दिया जाएगा और मुफ्त नौकरी दी जाएगी।"
]

# Write science fake samples
for idx, text in enumerate(science_fake_samples):
    file_path = f"{fake_folder}/science_fake_extra_{idx}.txt"
    with open(file_path, 'w', encoding='utf-8') as f:
        f.write(text)

# Write official-style fake samples
for idx, text in enumerate(official_fake_samples):
    file_path = f"{fake_folder}/official_fake_extra_{idx}.txt"
    with open(file_path, 'w', encoding='utf-8') as f:
        f.write(text)

print(f"Added {len(science_fake_samples)} science-based + {len(official_fake_samples)} official-style fake news samples to {fake_folder}")


# Load data
fake_df = load_text_files_from_folder(fake_folder, label=1)
real_df = load_text_files_from_folder(real_folder, label=0)

hindi_df = pd.concat([fake_df, real_df], ignore_index=True)
print(f"Dataset shape: {hindi_df.shape}")
print(hindi_df['label'].value_counts())
```

```python
# STEP 1: Import remaining required libraries
!pip install -q indic-nlp-library
from indicnlp.tokenize import indic_tokenize
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
import joblib
```

```python
# Hindi stopwords (manually defined)
hindi_stopwords = set("""
और क्या के को में कि है वह पर से यह जो थे था कर हो हैं लिए हम आप भी नहीं दिया इसलिए
कोई साथ कहा किया करना इसकी जैसे उनके कुछ बिना रहे रहा ऐसे यदि
""".strip().split())

# Combine with English stopwords
english_stopwords = set(stopwords.words('english'))
stop_words = hindi_stopwords.union(english_stopwords)
```

```python
stemmer = PorterStemmer()

def clean_text(text):
    # Keep Hindi + English letters only
    text = re.sub(r'[^a-zA-Z\u0900-\u097F\s]', ' ', str(text))

    # Tokenize
    tokens = list(indic_tokenize.trivial_tokenize(text))

    # Lowercase, remove stopwords and short tokens
    filtered = [t.lower() for t in tokens if t.lower() not in stop_words and len(t) > 1]

    # Stem English words (if any)
    stemmed = [stemmer.stem(w) if re.match('[a-zA-Z]+', w) else w for w in filtered]

    return ' '.join(stemmed)

# Apply cleaning
print("Cleaning text data...")
hindi_df['clean_text'] = hindi_df['text'].apply(clean_text)
```

```python
vectorizer = TfidfVectorizer(max_features=3000, ngram_range=(1,2))
X = vectorizer.fit_transform(hindi_df['clean_text'])
Y = hindi_df['label'].values
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=42)
```

```python
model = LogisticRegression(max_iter=300, solver='saga')
model.fit(X_train, Y_train)
```

```python
# Evaluate
train_preds = model.predict(X_train)
test_preds = model.predict(X_test)

print(f"\nTraining Accuracy: {accuracy_score(Y_train, train_preds):.4f}")
print(f"Test Accuracy: {accuracy_score(Y_test, test_preds):.4f}")

print("\nClassification Report (Test):")
print(classification_report(Y_test, test_preds, target_names=['Real', 'Fake']))
```

```python

# Save the model
joblib.dump(model, 'hindi_fake_news_model.pkl')

# Save the vectorizer
joblib.dump(vectorizer, 'tfidf_vectorizer.pkl')
```

```python
def predict_hindi_news(text, model=model, vectorizer=vectorizer, threshold=0.3):
    cleaned = clean_text(text)
    print("Cleaned text:", cleaned)
    vec = vectorizer.transform([cleaned])
    proba = model.predict_proba(vec)[0]
    print(f"Probability Real: {proba[0]:.4f}, Probability Fake: {proba[1]:.4f}")
    return "Fake" if proba[1] > threshold else "Real"
```

```python
# STEP 10: Example test
sample_text = """
सरकार ने घोषणा की है कि सभी नागरिकों को स्वास्थ्य बीमा मिलेगा, लेकिन योजना की पूरी जानकारी अभी सामने नहीं आई है।
विपक्ष ने इस घोषणा पर सवाल उठाए हैं और कहा है कि यह सिर्फ चुनावी स्टंट हो सकता है।
"""
print("Prediction:", predict_hindi_news(sample_text))
```
